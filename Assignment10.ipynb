```python
import pandas as pd
```

```python
df = pd.read_csv("trainKNN.txt", sep = ",", header=None)
```

```python
df
```

```python
df.columns = ['ID Number','RI','Sodium','Magnesium','Aluminium','Silicon','Pottasium','Calcium','Barium','Iron','Type_of_Glass']
```

```python
df
```

```python
df1=df.drop(['Type_of_Glass'],axis=1)
df1
```

```python
import numpy as np
```

```python
corr = df1.corr()
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.8:
            if columns[j]:
                columns[j] = False
selected_columns = df1.columns[columns]
df2 = df[selected_columns]
```

```python
df3=pd.read_csv('testKNN.txt',sep=",",header=None)
df3.columns = ['ID Number','RI','Sodium','Magnesium','Aluminium','Silicon','Pottasium','Calcium','Barium','Iron','Type_of_Glass']
```

```python
df3
```

```python
feature_cols = ['ID Number','RI','Sodium','Magnesium','Aluminium','Silicon','Pottasium','Barium','Iron']
x_test = df3[feature_cols]
y_test = df3.Type_of_Glass
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train = df2 # Features
y_train = df["Type_of_Glass"] # Target variable
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler();
X_train=sc_X.fit_transform(X_train);
x_test=sc_X.transform(x_test);
```

```python
knn = KNeighborsClassifier(n_neighbors=5,metric='seuclidean',p=2, algorithm='brute')
```

```python
knn = knn.fit(X_train,y_train)
```

```python
pred = knn.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
