**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
```

**Step 2: Importing dataset**

```python
df = pd.read_csv('Data (1).csv')
df
```

```python
df2 = df.groupby(['Country']).mean()
df2["Weight"] = df2['Salary']/df2['Age']
df2
```

**Step 3: Handling the missing data**

```python
b=0
for a in df['Salary']:
    if np.isnan(a) == True:
        df.loc[b,'Salary']=df2.loc[df.loc[b]['Country']]['Weight']*df.loc[b]['Age']      
    b=b+1
```

```python
df
```

```python
c=0
for a in df['Age']:
    if np.isnan(a) == True:
        df.loc[c,'Age']=df.loc[c]['Salary']/df2.loc[df.loc[c]['Country']]['Weight']    
    c=c+1
```

```python
df
```

**Step 4: Encoding categorical data**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Country_En']=le.fit_transform(df['Country'])
df
```

**Step 5: Creating a dummy variable**

```python
df1 = pd.get_dummies(df,columns=['Country'])
df1
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X=df[['Age','Salary','Country_En']]
y=df.Purchased
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=.2, train_size=.8, random_state=1, shuffle=True, stratify=None)
```

**Step 7: Feature Scaling**
